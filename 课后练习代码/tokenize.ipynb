```python
import os
path = r'D:\作业\NLP\bilingual'
filenames = os.listdir(path)
for filename in filenames:
    print(filename)
```

```python
alltext = ""
for filename in filenames:
    with open("D:/作业/NLP/bilingual/"+str(filename),encoding='utf-8') as f:
        text = f.read()
        #print(text)
        alltext = alltext+text
alltext
```

```python
import re
uncn = re.compile(r'[\u0061-\u007a,\u0020]') 
en = "".join(uncn.findall(alltext.lower()))
print(en)
```

```python
cn = re.compile(r'[\u4e00-\u9fa5]')
ch = "".join(cn.findall(alltext))
print(ch)
```

```python
from nltk import*
sent_tokenize(en)
```

```python
word_tokenize(en)
```

```python
pos_tag(word_tokenize(en))
```

```python
type(pos_tag(word_tokenize(en)))
```

```python
en_tag=pos_tag(word_tokenize(en))
len(en_tag)
```

```python
tags = []
for i in range(0,len(en_tag)):
    tags.append(en_tag[i][1])
tags
```

```python
tag_dict = {}
for tag in tags:
    if tag in tag_dict.keys():
        tag_dict[tag] = tag_dict[tag]+1
    else:
        tag_dict[tag] = 1
tag_dict
```

```python
tag_list=sorted(tag_dict.items(),key=lambda x:x[1],reverse=True)
```

```python
i = 0
for key in tag_list:
    print(key[0]+" "+str(key[1]))
    i+=1
    if(i==5):
        break
```

```python
from matplotlib import pyplot as plt
i=0
labels = []
size = []
for key in tag_list:
    labels.append(key[0])
    size.append(key[1])
    i+=1
    if(i==5):
        break
print(labels)
print(size)
```

```python
labels.append("others")
labels
```

```python
count_other = len(en_tag)
for count in size:
    count_other=count_other-count
count_other
```

```python
size.append(count_other)
size
```

```python
patches,l_text,p_text = plt.pie(size,labels=labels,
                                labeldistance = 1.1,autopct = '%3.1f%%',shadow = False,
                                startangle = 90,pctdistance = 0.6)

for t in l_text:
    t.set_size=(30)
for t in p_text:
    t.set_size=(20)

plt.axis('equal')
plt.legend()
plt.show()
```
