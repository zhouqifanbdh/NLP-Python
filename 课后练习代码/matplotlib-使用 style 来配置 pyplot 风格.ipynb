```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.style.available
```

```python
x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.plot(x, y)

plt.show()
```

```python
plt.style.use('ggplot')

plt.plot(x, y)

plt.show()
```

```python
with plt.style.context(('dark_background')):
    plt.plot(x, y, 'r-o')
    plt.show()
```

```python
with plt.style.context(('dark_background')):
    pass
plt.plot(x, y, 'r-o')
plt.show()
```

```python
plt.style.use(['dark_background', 'ggplot'])

plt.plot(x, y, 'r-o')
plt.show()
```
